```python
import random
import math
import matplotlib.pyplot as plt

def generar_normal():
    u1 = random.random()
    u2 = random.random()
    
    z1 = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    
    return z1

# Establecer la semilla
random.seed(83)

# Generar 10000 números aleatorios normales
numeros = [generar_normal() for _ in range(1000000)]

# Crear un histograma con 30 bins
plt.hist(numeros, bins=1000, density=True, edgecolor='blue')

# Mostrar el gráfico
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt


# Definir la media y la matriz de covarianza
media = [0, 0]
cov = [[1, 0.5], 
       [0.5, 1]]

# Generar 10000 muestras
muestras = np.random.multivariate_normal(media, cov, 10000)

# Extraer las componentes x e y de las muestras
x = muestras[:, 0]
y = muestras[:, 1]

# Crear un diagrama de dispersión de las muestras
plt.scatter(x, y, s=1)

# Mostrar el gráfico
plt.show()
```
