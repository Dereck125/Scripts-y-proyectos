```python
#CLASE ARTICULO PRECEDERO  la cual debera de tener (nombre del articulo, descripcion, fecha de caducidad)

class articulo_perecedero:
  def __init__(self, nombre, descripcion, caducidad,precio):
    self.nombre =  nombre
    self.descripcion = descripcion
    self.caducidad = caducidad
    self.precio = precio

  def get_precio(self):
    return self.precio
  def get_nombre(self):
    return self.nombre
  def get_caducidad(self):
    return self.caducidad 
  def caduca_1(self):
    if self.caducidad > 0:
      self.caducidad-=1
    else:
      return "El articulo caduco"
    return self.caducidad


  def __str__(self):
    return f"Nombre: {self.nombre} \nDescripcion:  {self.descripcion} \nCaducidad: {self.caducidad} dias \nPrecio: {self.precio}$"


```

Clase proveedor

```python
class proveedor:
  def __init__(self,nombre,fecha_de_nacimiento,rfc,gtotales,vdiarias,mermadiaria,ventastotales,mermatotal,articulos):
    self.nombre = nombre
    self.fecha_de_nacimiento = fecha_de_nacimiento
    self.rfc = rfc
    self.gtotales = gtotales
    self.vdiarias = vdiarias
    self.mermadiaria = mermadiaria
    self.ventastotales = ventastotales
    self.mermatotal = mermatotal
    self.articulos = articulos
  
  def __str__(self) -> str:
    return "PROVEEDOR \nNombre: {} \nRFC: {} \nVenta diaria: {} \nMerma diaria: {} \nGanancias totales: {} \nVentas totales:{} \nMerma total: {} \nLista de articulos y cantidad: {} ".format(self.nombre,self.rfc,self.vdiarias,self.mermadiaria,self.gtotales,self.ventastotales,self.mermatotal,self.articulos)

  def ganancias_totales(self):
    gtotales = self.ventastotales - self.mermatotal
    self.gtotales = gtotales
    return gtotales

  def get_articulos(self):
    return self.articulos

  def ganancias_diarias(self):
    gdiarias = self.vdiarias - self.mermadiaria
    return gdiarias

  def merma_diaria(self,v):
    self.mermadiaria += v
    return  self.mermadiaria 

  def merma_total(self,m):
      self.mermatotal+=m
      return self.mermatotal

  def get_nombre(self):
    return self.nombre

  def get_ventasT(self):
    return self.ventastotales

  def set_ventasT(self,vt):
    self.ventastotales+= vt
    return self.ventastotales


  def get_gananciasT(self):
    return self.gtotales

  def get_merma_diaria(self):
    return self.mermadiaria
  
  def get_mermaT(self):
    return self.mermatotal


  def vender_diario(self,cantidad_de_articulo):
    self.vdiarias+=cantidad_de_articulo
    return self.vdiarias

  def vender_total(self,vdiaria):
    self.ventastotales+=vdiaria
    return self.ventastotales

  def get_ventas_totales(self):
    return self.ventastotales
  
  def get_ventas_diarias(self):
    return self.vdiarias













#Obtener rfc
def obtener_rfc(nombre_c,fecha_n)->str:
    nombre_f = nombre_c
    fecha = fecha_n

    for i in range(6):
      nombre_f_l = nombre_c.split(" ")

      fecha_s_l = fecha_n.split("-")
      if i == 0:
        name_rfc = nombre_f_l[0]
        p_0 = name_rfc[0]
      elif i == 1:
        apellido_p = nombre_f_l[1]
        p_1 = apellido_p[0] + apellido_p[1]
      elif i == 2:
        apellido_m = nombre_f_l[2]
        p_2 = apellido_m[0]
      elif i== 3:
        anio = fecha_s_l[2]
        p_3 = anio[2]+ anio[3]
      elif i == 4:
        p_4 = fecha_s_l[1]
      elif i == 5:
        if int(fecha_s_l[0]) >=10:
         p_5 = fecha_s_l[0]
        else:
         p_5 = "0"+fecha_s_l[0]

      
    rfc_letras= p_1 + p_2 + p_0 + p_3 + p_4 + p_5
    return rfc_letras.upper()





                                
```

clase cliente

```python
class cliente:
  def __init__(self,nombre,n_cliente,pdiario,pgastado,articulos_comprados):
    self.nombre= nombre
    self.n_cliente= n_cliente
    self.pdiario=pdiario
    self.pgastado=pgastado
    self.articulos_comprados = articulos_comprados

  def get_pdiario(self):
    return self.pdiario
  def get_nombre(self):
    return self.nombre
  def get_articulos(self):
    return self.articulos_comprados
  def nuevo_dia(self,extra):
    self.pdiario+=extra
    return self.pdiario
    
  def __str__(self) -> str:
    return "CLIENTE  \nNumero: {} \nNombre: {} \nPresupuesto actual: {} \nPresupuesto gastado: {} \nArticulos comprados: {}".format(self.n_cliente,self.nombre,self.pdiario,self.pgastado,self.articulos_comprados)

  def compra(self,precio,cantidad):
    if self.pdiario >= self.pgastado and self.pdiario > (precio*cantidad):
      self.pdiario = self.pdiario - (precio*cantidad)
      self.pgastado+= (precio*cantidad)
    
    return self.pgastado, self.pdiario

  def si_compra(self,precio,cantidad):
    if self.pdiario >= self.pgastado and self.pdiario > (precio*cantidad):
      return True

  
```
