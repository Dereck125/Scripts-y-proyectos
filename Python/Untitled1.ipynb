```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

#Suposicion de que todos los clientes pagan 252 pesos

```python
# Datos de entrada
X = np.array([[21], [17], [4]])
y = np.array([5300,4284,1008])
```

```python

# Crear un objeto de regresión lineal
regresion_lineal = LinearRegression()

# Entrenar el modelo utilizando los datos
regresion_lineal.fit(X, y)

# Predecir los valores de salida
y_pred = regresion_lineal.predict(X)

# Imprimir los coeficientes de la regresión lineal
print("Coeficientes: ", regresion_lineal.coef_)
print("Intercepto: ", regresion_lineal.intercept_)

# Imprimir las predicciones
print("Valores predichos: ", y_pred)
```

```python
import matplotlib.pyplot as plt

# Graficar los puntos de datos
plt.scatter(X[:, 0], y, color='blue', label='Datos reales')

# Graficar la línea de regresión
plt.plot(X[:, 0], y_pred, color='red', label='Regresión lineal')

# Configurar etiquetas y leyenda
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión lineal')
plt.legend()

# Mostrar la gráfica
plt.show()

print("  ")
# Valor para hacer la predicción
nuevo_valor = np.array([[21]])

# Hacer la predicción
prediccion = regresion_lineal.predict(nuevo_valor)

# Imprimir el resultado de la predicción
print("Predicción:", prediccion)
bruto = prediccion[0] - 3200
print(bruto)

print("50-50")
print(f"Katy: {bruto*0.5}, Jesus {bruto * 0.5}")
print("53-47")
print(f"Katy: {bruto*0.53}, Jesus {bruto * 0.47}")
print("55-45")
print(f"Katy: {bruto*0.55}, Jesus {bruto * 0.45}")
print("60-40")
print(f"Katy: {bruto*0.6}, Jesus {bruto * 0.4}")
print("80-20")
print(f"Katy: {bruto*0.8}, Jesus {bruto * 0.2}")
```
