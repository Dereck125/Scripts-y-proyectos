```python
#Se lee cada linea del archico txt, par asi extraer cada password
with gzip.open("/content/drive/MyDrive/TareaDrive/Lista_de_contrasenias.gz",'rb') as contrasenias:
    lista_contrasenias = contrasenias.readlines()
    #Hacemos la variable de tipo list de la lista de passwords
    list(lista_contrasenias)
```

```python
#Como cada elemento de la lista es de tipo byte y lo nececitamos str, 
#decodificamos su formato para pasarlo a str, cada elemento de la lsita

#Variable para la nueva lista
nueva_lista_contra = []
for i in range(len(lista_contrasenias)):
  #se decodifica cada elemnto
  nueva_lista_contra.append(lista_contrasenias[i].decode('UTF-8'))
```

#Datos estadisticos


```python
#variables que contienen los caracteres que queremos comparar de cada password

letras_min = "qwertyuiopasdfghjklzxcvbnm"
letras_may = "QWERTYUIOPASDFGHJKLZXCVBNM"
numeros = "1234567890"

#Primero extraemos cada caracter inicial de cada palabra en la lista,
#creamos una nueva variable para guardar estos caracteres
caracteres_iniciales =[]
#Se revisa cada password y en ella su elemento 0, es decir el caracter inicial

for password in nueva_lista_contra:
  caracteres_iniciales.append(password[0])

```

```python
#Lista de los primeros caracteres de cada password
print(caracteres_iniciales)
```

```python
#Creamos las variables que van a contar la cantidad de veces en que encuentra una cincidencia

c_letras_may= 0
c_letras_min= 0
c_numeros = 0

#Listas con los caracteres buscados, para realizar la comparacion
comparacion_1=[x for x in letras_may ]
comparacion_2=[i for i in letras_min]
comparacion_3=[j for j in numeros]

for i in range(len(caracteres_iniciales)):
  for j in range(len(comparacion_1)):
    if caracteres_iniciales[i] == comparacion_1[j]:
      c_letras_may+=1
  for k in range(len(comparacion_2)):
    if caracteres_iniciales[i] == comparacion_2[k]:
      c_letras_min+=1
  for l in range(len(comparacion_3)):
    if caracteres_iniciales[i] == comparacion_3[l]:
      c_numeros+=1
  

```

```python
#Cantidad de otros caracteres, seran los que no fueron contados anteriormente
otros = len(caracteres_iniciales) - c_letras_may - c_letras_min - c_numeros
```

```python
#Estadistica final de la lista de passwords

print("-"*70)
print(" ")
print("Nombre del archivo es: Top304Thousand-probable-v2.txt")
print("Las Password de WiFi mas repetidas")
print(f"La cantidad de elementos de la lista es :{len(nueva_lista_contra)}")

print("-"*70)

print(f"Cantidad de passwords que empiezan con: \n-Letras mayusculas: {c_letras_may}", 
      f"\n-Letras minusculas : {c_letras_min}",f"\n-Numeros: {c_numeros}",
      f"\n-Otros caracteres: {otros}")
print("--"*70)
print(f"Total: {c_letras_may + c_letras_min + c_numeros + otros}")
print("--"*70)
print(f"Es decir:\n {round((c_letras_may*100)/len(nueva_lista_contra),2)}% inicia con letra mayuscula \n {round((c_letras_min*100)/len(nueva_lista_contra),2)}% inicia con letra minuscula\n {round((c_numeros*100)/len(nueva_lista_contra),2)}% inicia con un numero\n {round((otros*100)/len(nueva_lista_contra),2)}% inicia con otro caracter")
```

```python
cantidades_de_password =[c_numeros,c_letras_may,c_letras_min,otros]
tipos_de_password =["Numeros","Mayusculas","Minusculas","Otros"]
colores = ["#EE6055","#60D394","#AAF683","#FFD97D","#FF9B85"]
plt.title("")
plt.pie(cantidades_de_password,labels=tipos_de_password, autopct="%0.02f %%", colors=colores)
plt.axis("equal")
plt.show()
```
