3. ¿Que sexo es el que tiene mas
cuentas de inversion? 

```python
@interact(inversion =["Si","No"], sexo=["F","M"])
def consulta3(inversion="Si",sexo="F"):
  Consulta= f'''
  SELECT sexo ||' FEMENINO: '|| count(sexo)
  FROM [ALUMNOS] JOIN [INVERSION]
  ON [ALUMNOS].id_alumnos = [INVERSION].id_alumnos
  WHERE inversion = "{inversion}" AND sexo ="{sexo}" ;
  '''
  cursor.execute(Consulta)
  return imprimirCursor(cursor)
 
```
