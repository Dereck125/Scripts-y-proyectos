```python
import random as rnd
import time
from datetime import timedelta
import threading
```

```python
tipos_de_cuenta = {1:"Debito",2:"Credito",3:"No-cuentahabiente"}
class cuenta_habiente:
  def __init__(self,nombre,direccion,telefono,NC,tipoc):
    self.nombre = nombre
    self.direccion = direccion
    self.telefono = telefono
    self.numero_de_cuenta = NC
    self.tipoc = tipoc
  
  def show_data(self):
    print("-"*30)
    print(f"Nombre del cliente: {self.nombre}\nDireccion: {self.direccion}\nTelefono: {self.telefono}\nNumero de cuenta: {self.numero_de_cuenta}\nTipo de cuenta: {self.tipoc}")
    print("-"*30)
```

```python
class cola_de_banco:

    def __init__(self):
        self.cola=[]

    def enq(self,x):
        self.cola.insert(0,x)

    def deq(self):
        assert len(self.cola)>0
        return self.cola.pop()

    def is_empty(self):
        return len(self.cola)==0
```

```python
class cajero:

    def __init__(self):
        self.cola=[]

    def enq(self,x):
        self.cola.insert(0,x)

    def deq(self):
        assert len(self.cola)>0
        return self.cola.pop()

    def is_empty(self):
        return len(self.cola)==0
```

```python
#Variables para generar los clientes aleatoriamente
nombres ="Alma,Arely,Andrea,Alondra,Blanca,Bruno,Brandon,Cesia,Carlos,Citlalí,Coraline,Daniela,Dilan,Denis,Esmeralda,Emiliano,Estér,Eliseo,Ellie"

apellidos = "Mendez,Morales,Mojica,Martínez,Niño,Olmeda,Ortega,Ortiz,Puente,Pérez,Pineda,Peña,Piña,Pardo,Rosales,Rivera,Ramírez,Ruiz,Rodríguez"

avca=["AV.","CALLE"]

numero_cliente = []
nombre_completo= []
direcciones = []
lista_de_clientes =[] 
cola_de_clientes = cola_de_banco()


#Cajeros
cajero_1= cajero()
cajero_2= cajero()
cajero_3=cajero()



#Tiempo que tardo en atender cada caja
t_1= 0
t_2 = 0
t_3 = 0

long = 0

m=int(input("Cuantos minutos deseas emular el Banco: "))

c_1=0
c_2=0
c_3=0


f=1

#Generamos nombre aleatorios
for i in range(1,110):
   x = nombres.split(",")[rnd.randrange(1,19)] + " "+  apellidos.split(",")[rnd.randrange(1,19)] + " "+  apellidos.split(",")[rnd.randrange(1,19)]
   nombre_completo.append(x)

  #Numero de cliente
for j in range(1,110):
    y =f"CBN{rnd.randrange(100,992)}"
    numero_cliente.append(y)

  #Direccion
for i in range(1,110):
   d = avca[rnd.randrange(1,2)] + " " + nombres.split(",")[rnd.randrange(1,19)].upper() + " " + "NO. " + str(rnd.randrange(1,300)) 
   direcciones.append(d)

  # Clientes aleatorios
for k in range(1,110):
  cliente = cuenta_habiente(nombre_completo[rnd.randrange(1,100)],direcciones[rnd.randrange(1,100)],rnd.randrange(5500000000,5599999999),numero_cliente[rnd.randrange(1,100)],tipos_de_cuenta.get(rnd.randrange(1,4)))
  lista_de_clientes.append(cliente)
```

```python
#Funcion que crea a los clientes cada minuto
def crear_clientes(cola_de_clientes,m,lista_de_clientes):
    #Contadores de clientes por caja
    while cola_de_clientes.is_empty() and m>0:
        m-=1 
        if  m>0:    
         
         print("Llegaron nuevos clientes")
         
        else:

            
            print("-"*30)
            print("Llegaron los ultimos clientes, se atenderan y el banco cerrara")
            print("-"*30)
            
         
        for k in range(rnd.randrange(0,10)):
            V= rnd.randrange(1,100)
            print(f"El cliente {lista_de_clientes[V].nombre} acaba de entrar y es de tipo {lista_de_clientes[V].tipoc}")
            
            cola_de_clientes.enq(lista_de_clientes[V])

            if lista_de_clientes[V].tipoc == "Debito":

                
                cajero_1.enq(lista_de_clientes[V])
                cola_de_clientes.deq()

            if lista_de_clientes[V].tipoc == "Credito":

                
                cajero_2.enq(lista_de_clientes[V])
                cola_de_clientes.deq()

            if lista_de_clientes[V].tipoc == "No-cuentahabiente":

                
                cajero_3.enq(lista_de_clientes[V])
                cola_de_clientes.deq() 
        
        time.sleep(60) 
```

```python
inicio_banco = time.time()
print("-"*80)
print("         EL BANCO ABRE SUS CAJAS Y INICA LAS OPERACIONES          ")
print("-"*80)
#Cada minuto llegan entre 0 o 10 nuevos clientes a la fila 
hilo = threading.Thread(target=crear_clientes, args=(cola_de_clientes,m,lista_de_clientes, ))
hilo.start()


for i in range(m):
   
   inicio_1 = time.time()
   while cajero_1.is_empty() == False:
      c_1+=1
      cajero_1.deq()
      print(f"Hay {len(cajero_1.cola)+1} formados en el cajero 1")
      
      time.sleep(rnd.randrange(5,60))
   fin_1 = time.time()
   t_1+=abs(fin_1-inicio_1)
   print(f"Caja para cuentas Debito Libre, tardo {timedelta(seconds=abs(fin_1-inicio_1))}, en atender a todos ")

   
   inicio_2 = time.time()
   while cajero_2.is_empty() == False:
      c_2+=1
      cajero_2.deq()
      print(f"Hay {len(cajero_2.cola)+1} formados en el cajero 2")
      
      time.sleep(rnd.randrange(5,60))
   fin_2 = time.time()
   t_2+=abs(fin_2-inicio_2)
   print(f"Caja pa cuentas Credito Libre, tardo {timedelta(seconds=abs(fin_2-inicio_2))}, en atender a todos  ")

   inicio_3 = time.time()
   while cajero_3.is_empty() == False:
      c_3+=1
      cajero_3.deq()
      print(f"Hay {len(cajero_3.cola)+1} formados en el cajero 3")
      
      time.sleep(rnd.randrange(5,60))
   fin_3 = time.time()
   t_3+=abs(fin_3-inicio_3)
   print(f"Caja para No-cuentahabientes Libre, tardo {timedelta(seconds=abs(fin_3-inicio_3))}, en atender a todos ")

final_banco = time.time()


tiem_total=timedelta(seconds= abs(( (final_banco-inicio_banco)-(m*60))))


if cajero_1.is_empty and cajero_2.is_empty and cajero_3.is_empty and cola_de_clientes.is_empty:
   
   print("-"*50)
   if c_1 !=0:
      prom_1=timedelta(seconds=t_1/c_1)
      print(f"El cajero 1, en prmedio tarda {prom_1} en atender a cada cliente y pasaron {c_1} clientes")
      print("  ")
   if c_2 !=0:
      prom_2=timedelta(seconds=t_2/c_2)
      print(f"El cajero 2, en prmedio tarda {prom_2}  en atender a cada cliente y pasaron {c_2} clientes")
      print("  ")
   if c_3 !=0:
      prom_3=timedelta(seconds=t_3/c_3)
      print(f"El cajero 3, en prmedio tarda {prom_3} en atender a cada cliente y pasaron {c_3} clientes")

   print("  ")
   print(f"El banco atendio a todos y tardo {tiem_total} mas en cerrar, atendiendo en total a {c_1+c_2+c_3} clientes")
   print("-"*50)
```
